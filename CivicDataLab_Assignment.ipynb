## Data Preprocessing

```python
#Separating data from df consisting of primary education only
primary_school_df = df[df['Grant Head Description'] == 'शिक्षा विभाग (प्राथमिक शिक्षा)']
```

```python
primary_school_df.shape
```

```python
primary_school_df['Major Head Code'].unique()
```

```python
primary_school_df['Major Head Description'].unique()
```

```python
primary_school_df = primary_school_df[primary_school_df['Major Head Description'] == 'सामान्य शिक्षा']
```

```python
primary_school_df.shape
```

```python
primary_school_df['Major Head Code'].unique()
```

```python
pd.set_option('display.max_columns', None)
display(primary_school_df.sample(10))
```

```python
# Separating data consisting of secondary education only
secondary_school_df = df[df['Grant Head Description'] == 'शिक्षा विभाग (माध्यमिक शिक्षा)']
```

```python
secondary_school_df.shape
```

```python
secondary_school_df['Major Head Code'].unique()
```

```python
secondary_school_df['Major Head Description'].unique()
```

```python
secondary_school_df = secondary_school_df[secondary_school_df['Major Head Description'] == 'सामान्य शिक्षा']
```

```python
secondary_school_df.shape
```

```python
secondary_school_df['Major Head Code'].unique()
```

```python
secondary_school_df.sample(10)
```

```python
# concatening primary and secondary school education to get our required dataframe
school_df = pd.concat([primary_school_df, secondary_school_df])
```
